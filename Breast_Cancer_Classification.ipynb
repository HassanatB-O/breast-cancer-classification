Data pre-processing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
# loading the dataset
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
# adding the target column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
data_frame.tail()
```

```python
data_frame.shape
```

```python
data_frame.isnull().sum()
```

```python
data_frame['label'].value_counts()
```

1 -> Benign
0 -> Malignant

```python
data_frame.groupby('label').mean()
```

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

Splitting the data into training & test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X_train.shape, X.shape, X_test.shape)
```

Standardize the Data cos of the accuracy

```python
print(X_train)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

Building the Neural Network

```python
# importing tensorflow and keras for the neural network
import tensorflow as tf
tf.random.set_seed(3) #setting it makes the accuracy score the same everytime the code is run
from tensorflow import keras
```

```python
# setting up the layers of the neural network

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)), # Flatten conerts the data nto a singe dimensional array
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the neural network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc='upper right')
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

```python
Y_pred = model.predict(X_test_std)
```

```python
print(Y_pred.shape)
print(Y_pred[0])
```

model.predict() gives the prediction probability of each class for that data point

```python
# to convert the Y_pred into class labels, (0, 1) not just a probablity
# meaning if the first value in the array = 0, the label = 0 and vice versa
Y_pred_labels = [int(np.argmax(i)) for i in Y_pred]
print(Y_pred_labels)
```

Building the Predictive System

```python
input_data = (13.54,14.36,87.46,566.3,0.09779,0.08129,0.06664,0.04781,0.1885,0.05766,0.2699,0.7886,2.058,23.56,0.008462,0.0146,0.02387,0.01315,0.0198,0.0023,15.11,19.26,99.7,711.2,0.144,0.1773,0.239,0.1288,0.2977,0.07259)

# change the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array since we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# Standardize the data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

# converting to class label
prediction_label = [int(np.argmax(prediction))]
print(prediction_label)

if prediction_label[0] == 0:
  print("The tumor is Malignant")
else:
  print("The tumor is Benign")
```
